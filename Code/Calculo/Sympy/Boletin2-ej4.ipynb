```python
import sympy as sp

x = sp.Symbol('x', real=True)
_f = sp.sqrt(x**2 + 1) - x

f = sp.Lambda(x, _f)

print('Verticales:')
print('Singularidades = ', sp.calculus.singularities(f(x), x))

print('Horizontales:')
print("lim x -> -oo=", sp.limit(f(x), x, -sp.oo))
print("lim x -> oo=",sp.limit(f(x), x, sp.oo))

print('Oblicuas:')
print('m = ', sp.limit(f(x) / x, x, -sp.oo))
print('n')

sp.plot(f(x))
```
