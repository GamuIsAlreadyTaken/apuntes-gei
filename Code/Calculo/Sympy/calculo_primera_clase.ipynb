```python

x = sp.Symbol('x', real=True)

f1_expr = sp.cos(x)
f2_expr = (x**2) / (x+1)

f = sp.Lambda(x, f1_expr + (f2_expr - f1_expr)*sp.Heaviside(x, 0))

display(sp.limit(f(x), x, -1, dir="+") if sp.limit(f(x), x, -1, dir="-") == sp.limit(f(x), x, -1, dir="+") else 'No existe el limite')

display(sp.limit(f(x), x, 1, dir="-") if sp.limit(f(x), x, 1, dir="+") == sp.limit(f(x), x, 1, dir="-") else 'No existe el limite')
display(sp.limit(f(x), x, 0, dir="+") if sp.limit(f(x), x, 0, dir="-") == sp.limit(f(x), x, 0, dir="+") else 'No existe el limite')

sp.plot(f(x))

```

```python
x = sp.Symbol('x', real=True)

f_ = sp.Abs(2*x+4)-sp.Abs(5*x-2)

f = sp.Lambda(x, f_)
print(sp.solve(f_, x))

sp.plot(f(x))
```

```python

x, y = sp.symbols('x,y', real=True)
f_ = sp.exp(2*x)

finv_ = sp.solve(y - f_, x)[0].subs({y:x})

f = sp.Lambda(x, f_)
finv = sp.Lambda(x, finv_)


print(f(finv).simplify())
sp.plot(f(finv(x)), f(x), finv(x), (x, -10, 10), ylim=(-10, 10), legend=True)


```
