```python
import sympy as sp
x = sp.Symbol('x', real=True)
```

```python
f_ = sp.Abs(x**2 - 8)
f = sp.Lambda(x, f_)

print('Par: {}'.format(f(x) == f(-x)))
print('Inpar: {}'.format(f(x) == -f(-x)))

sp.plot_implicit(f(x) < 10, (x, -20, 20))
```

```python
t, y = sp.symbols('t, y', real=True)
g1_ = (sp.cos(t) + t) ** 2
g2_ = (t**3 + 1) / (t**2 - 4)

g1 = sp.Lambda(t, g1_)
g2 = sp.Lambda(t, g2_)

g = sp.Lambda(t, g1_ + (g2_ - g1_) * sp.Heaviside(t - 2*sp.pi, 0))

display(sp.limit(g(t), t, 2*sp.pi, dir='+'), sp.limit(g(t), t, 2*sp.pi, dir='-'))
print('Dominio = todo x, porque las singularidades {} no pertenecen a la funcion en ese tramo '.format(sp.calculus.singularities(g(t), t)))
# _2 = sp.Lambda(y, 2)
plot = sp.plot(g(t), (t, -10, 20), ylim = (0, 60))
```
