```python
import sympy as sp

# Queremos una función, un intervalo
# Calcular el punto de corte de la tangente con y=0
# actualizar raiz

x = sp.Symbol('x', real=True)

f_ = x**2 - 1
interval = [-2, 2]
root = (interval[0] + interval[1]) / 2 

maxIter = 100
tol = 10 ** -9

f_prim_ = sp.diff(f_, x)
f = sp.Lambda(x, f_)
f_prim = sp.Lambda(x, f_prim_)
for i in range(1,maxIter):
  if(f_prim(root) == 0): root +=tol
  root -= f(root) / f_prim(root)

display(root)

sp.plot(f(x), xlim=(interval), ylim=(-2,2))

```
